```python

from sklearn.datasets import load_iris

# Cargar el dataset Iris
iris = load_iris()

```

```python
# Mostrar la descripción del dataset
print(iris.DESCR)

# Ver las características del dataset
print("Características del dataset:")
print(iris.feature_names)
```

```python
import matplotlib.pyplot as plt

# Graficar un diagrama de dispersión para las primeras dos características
plt.figure(figsize=(8, 6))
plt.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title('Diagrama de dispersión de las dos primeras características')
plt.colorbar(label='Clase', ticks=[0, 1, 2], format=plt.FuncFormatter(lambda x, _: iris.target_names[int(x)]))
plt.show()
```
